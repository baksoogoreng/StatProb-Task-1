**Deskripsi Statistik**

```python
import pandas as pd
import numpy as np
from scipy import stats

# Membaca file Excel dan memilih sheet 'Sheet1'
df = pd.read_excel('Dataset.xlsx', sheet_name='Sheet1')

# Mengambil kolom yang ingin dianalisis dan menghilangkan baris dengan NaN
data = df[['Curah Hujan (mm)', 'Penjualan A (pcs)', 'Penjualan B (pcs)']].dropna()

# Loop untuk menghitung statistik pada setiap kolom
for column in data:
    print(f"\nStatistik untuk kolom: {column}")
    
    # Mengambil data dari kolom yang sedang diproses
    col_data = data[column]
    
    # Menghitung statistik
    mean = np.mean(col_data)
    median = np.median(col_data)
    mode_result = stats.mode(col_data, keepdims=True)
    
    if mode_result.count.size > 0:
        mode = mode_result.mode[0]
    else:
        mode = 0
    
    range_data = np.ptp(col_data)
    iqr = stats.iqr(col_data)
    variance = np.var(col_data, ddof=1)
    std_dev = np.std(col_data, ddof=1)
    skewness = stats.skew(col_data)
    kurtosis = stats.kurtosis(col_data)
    
    # Menampilkan hasil per kolom
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Range: {range_data}")
    print(f"IQR: {iqr}")
    print(f"Varians: {variance}")
    print(f"Standar Deviasi: {std_dev}")
    print(f"Skewness: {skewness}")
    print(f"Kurtosis: {kurtosis}")
```

**Visualisasi**


```python
import matplotlib.pyplot as plt


df = pd.read_excel('Dataset.xlsx', sheet_name='Sheet1')

data = df['Penjualan B (pcs)'].dropna()

plt.figure(figsize=(8, 6))
plt.hist(data)
plt.title("Histogram Penjualan B")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()

bins = np.arange(min(data), max(data), 10)
frequency_table = pd.cut(data, bins=bins).value_counts().sort_index()
print("Tabel Distribusi Frekuensi:")
print(frequency_table)

midpoints = 0.5 * (bins[1:] + bins[:-1])
frequencies = frequency_table.values
plt.figure(figsize=(10, 6))
plt.plot(midpoints, frequencies, marker='o', linestyle='-', color='b')  # Membuat poligon frekuensi
plt.title("Poligon Frekuensi Penjualan B")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.grid(True)
plt.show()
```

**Boxplot**

Boxplot adalah grafik yang merangkum data dengan menampilkan nilai terkecil, kuartil, median, dan outlier. Histogram menunjukkan berapa kali nilai-nilai dalam data muncul, dengan membaginya ke dalam kelompok tertentu. Jadi, boxplot digunakan untuk melihat ringkasan dan outlier, sedangkan histogram fokus pada pola sebaran data dan frekuensi kemunculannya.

**Outlier**


Outlier adalah nilai atau observasi yang berbeda secara signifikan dari nilai-nilai lain dalam kumpulan data. Outlier bisa berupa nilai yang sangat tinggi atau rendah, atau bahkan keluar dari pola yang diharapkan dalam data, dapat mencakup nilai maksimum sampel atau nilai minimum sampel , atau keduanya, tergantung pada apakah nilai tersebut sangat tinggi atau rendah.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data dari Excel (asumsikan sheet_name jika tidak diperlukan, dihapus)
df = pd.read_excel('Dataset.xlsx')

# Asumsikan file memiliki kolom 'Curah Hujan (mm)', 'Penjualan A (pcs)', dan 'Penjualan B (pcs)'
hujan = df['Curah Hujan (mm)']  # Kolom curah hujan (mm)
penjualan_a = df['Penjualan A (pcs)']  # Kolom penjualan produk A
penjualan_b = df['Penjualan B (pcs)']  # Kolom penjualan produk B

# Membuat scatter plot untuk Penjualan A dan Penjualan B
plt.scatter(hujan, penjualan_a, color='red', label='Penjualan A')
plt.scatter(hujan, penjualan_b, color='yellow', label='Penjualan B')

# Menambahkan title, labels, dan legend
plt.title('Pengaruh Curah Hujan terhadap Penjualan A dan B')
plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Penjualan (pcs)')
plt.grid(True)
plt.legend()  # Untuk menampilkan label 'Penjualan A' dan 'Penjualan B'

# Tampilkan plot
plt.show()

# Menampilkan data dalam bentuk tabel di console
print(df)

# Korelasi antara curah hujan dan penjualan A
correlation_a = hujan.corr(penjualan_a)
print(f'Korelasi antara curah hujan dan penjualan A: {correlation_a}')

# Korelasi antara curah hujan dan penjualan B
correlation_b = hujan.corr(penjualan_b)
print(f'Korelasi antara curah hujan dan penjualan B: {correlation_b}')
```
